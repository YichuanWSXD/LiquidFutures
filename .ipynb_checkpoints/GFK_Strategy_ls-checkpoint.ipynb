```python
#!/usr/bin/python
# -*- coding: utf-8 -*-

# mac.py
import statsmodels.api as sm
from __future__ import print_function
import math 
import datetime
import seaborn as sns
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pylab as plt
from strategy import Strategy
from event import SignalEvent
from backtest import Backtest
from data import HistoricCSVDataHandler
from execution import SimulatedExecutionHandler
from portfolio_to_position import Portfolio
import datetime



class Treasury(Strategy):
    """
    
    """

    def __init__(
        self, bars, events, position,holding_days,num_pick
    ):
        """
        
        """
        self.bars = bars
        self.symbol_list = self.bars.symbol_list
        self.events = events
        self.position = position
        self.holding_days = holding_days
        self.position_index = list(self.position.index)
        self.clear_date = datetime.datetime(2099,1,1)
        # Set to True if a symbol is in the market
        self.bought = self._calculate_initial_bought()
        self.next_trading_date = self.position_index.pop(0)
        self.num_pick = num_pick
        self.change_date = 0
        self.bar_date = None
        

    def _calculate_initial_bought(self):
        """
        Adds keys to the bought dictionary for all symbols
        and sets them to 'OUT'.
        """
        bought = {}
        for s in self.symbol_list:
            bought[s] = 'OUT'
        return bought

    def calculate_signals(self, event):
        """
  

        Parameters
        event - A MarketEvent object. 
        """
        if event.type == 'MARKET':
            bar_date = self.bars.get_latest_bar_datetime(self.symbol_list[0])
            self.bar_date=bar_date
            if self.holding_days ==90:
                self.clear_date = datetime.datetime(2999,1,1)
            
            
            if bar_date >= self.clear_date and bar_date<self.next_trading_date:
                
                for s in self.symbol_list:
                    bars = self.bars.get_latest_bars_values(
                        s, "adj_close"
                    )
                    if bars is not None and not math.isnan(bars):
                        symbol = s
                        dt = datetime.datetime.utcnow()
                        sig_dir = ""

                        if  self.bought[s] == "LONG":
                            #print("SELL: %s" % bar_date)
                            sig_dir = 'EXIT'
                            signal = SignalEvent(1, symbol, dt, sig_dir, 1.0)
                            self.events.put(signal)
                            self.bought[s] = 'OUT'
                            
                        elif self.bought[s] == "SHORT":
                            #print("COVER: %s" % bar_date)
                            sig_dir = 'EXIT'
                            signal = SignalEvent(1, symbol, dt, sig_dir, 1.0)
                            self.events.put(signal)
                            self.bought[s] = 'OUT'

                        
                    self.clear_date = datetime.datetime(2099,1,1)
            
            bars = self.bars.get_latest_bars_values(self.symbol_list[0], "adj_close")        
            if bar_date >= self.next_trading_date and not math.isnan(bars):
                for s in self.symbol_list:
                    
                    bars = self.bars.get_latest_bars_values(
                        s, "adj_close"
                    )
                    
                    if bars is not None and not math.isnan(bars):
                        self.change_date = 1
#                         print(bars)
                        position_now = self.position.loc[self.next_trading_date,s]
                        symbol = s
                        dt = datetime.datetime.utcnow()
                        sig_dir = ""

                        if position_now==1:
                            #print("LONG: %s" % bar_date)
                            sig_dir = 'LONG'
                            signal = SignalEvent(1, symbol, dt, sig_dir, 1.3/self.num_pick)
                            self.events.put(signal)
                            self.bought[s] = 'LONG'
                            
                        elif position_now==-1:
                            #print("SHORT: %s" % bar_date)
                            sig_dir = 'SHORT'
                            signal = SignalEvent(1, symbol, dt, sig_dir, 0.3/self.num_pick)
                            self.events.put(signal)
                            self.bought[s] = 'SHORT'
                        elif position_now==0 and self.bought[s]!='OUT':
                            #print("LONG: %s" % bar_date)
                            sig_dir = 'EXIT'
                            signal = SignalEvent(1, symbol, dt, sig_dir, 0.05)
                            self.events.put(signal)
                            self.bought[s] = 'OUT'
                        
                if self.position_index != []:
                    self.next_trading_date = self.position_index.pop(0)
                    self.clear_date = bar_date+datetime.timedelta(self.holding_days)
                else:
                    self.next_trading_date = datetime.datetime(2999,1,1) 
                
                            
                    
                    

if __name__ == "__main__":

    equity_dict = dict() #key: num_pick + '_' + holding_2days
    data = pd.read_csv('../../input/long_short_stock.csv',index_col='Date',parse_dates=['Date'])
    position = pd.read_csv('../../input/position_Y.csv',index_col='PERIOD',parse_dates=['PERIOD'])
#     position.index += datetime.timedelta(30)
    position_new = position.copy()
    rank_high = pd.read_csv('../../input/long_short_rank/df_rank_high_Z.csv',index_col='PERIOD',parse_dates=['PERIOD'])
    rank_low = pd.read_csv('../../input/long_short_rank/df_rank_low_Z.csv',index_col='PERIOD',parse_dates=['PERIOD'])
#     rank_high.index += datetime.timedelta(30)
#     rank_low.index += datetime.timedelta(30)
    returns = pd.DataFrame(np.zeros([len(num),len(days)]),index = num,columns=days)
    sharpe = pd.DataFrame(np.zeros([len(num),len(days)]),index = num,columns=days)
    max_drawdown = pd.DataFrame(np.zeros([len(num),len(days)]),index = num,columns=days)
    drawdown_duration = pd.DataFrame(np.zeros([len(num),len(days)]),index = num,columns=days)
    for i in range(len(num)): 
        
        num_pick = num[i]
        position_new.loc[:,:] = 0 
        position_new[rank_high<=num_pick]=1
        position_new[rank_low<=num_pick]=-1
        for j in range(len(days)):
            holding_days = days[j]
            csv_dir = '../../input/long_short_stock/'
            symbol_list = list(position_new.columns)
            initial_capital = 1000000.0
            heartbeat = 0.0
            start_date = datetime.datetime(2009, 4, 1, 0, 0, 0)
#             end_date = datetime.datetime(2012,4,14,0,0,0)
            
            
            
#             start_date = end_date
            end_date = datetime.datetime(2019,10,16,0,0,0)
            position_new=position_new[position_new.index>start_date]
            print(f'{num_pick} stocks {holding_days} holding days:')
            backtest = Backtest(
                csv_dir, symbol_list,initial_capital, heartbeat, 
                start_date, end_date, HistoricCSVDataHandler, SimulatedExecutionHandler, 
                Portfolio, GfK_long_short,position_new,holding_days,num_pick
            )

            equity = backtest.simulate_trading();
            equity_dict[f'{num_pick}_{holding_days}']=equity
            stats = backtest.stats
            returns.loc[num_pick,holding_days]=stats[0][1]
            sharpe.loc[num_pick,holding_days]=stats[1][1]
            max_drawdown.loc[num_pick,holding_days]=stats[2][1]
            drawdown_duration.loc[num_pick,holding_days]=stats[3][1]
            print('-----------------------------------------------------\n\n')
            
            
            
            
            
            
#             plt.subplot(331+i)
#             equity_dict[holding_days]=equity
#             equity.total.plot()
#             plt.title(f'{days[i]} holding days')
#             plt.tight_layout()

#     data = pd.read_csv('../../input/long_short_stock.csv',index_col='Date',parse_dates=['Date'])
#     position = pd.read_csv('../../input/position_Y.csv',index_col='PERIOD',parse_dates=['PERIOD'])
#     holding_days = 60
#     csv_dir = '../../input/long_short_stock/'
#     symbol_list = list(data.columns)
#     initial_capital = 1000000.0
#     heartbeat = 0.0
#     start_date = datetime.datetime(2008, 1, 1, 0, 0, 0)
#     print(f'60 holding days:')
#     backtest = Backtest(
#         csv_dir, symbol_list, position,holding_days,initial_capital, heartbeat, 
#         start_date, HistoricCSVDataHandler, SimulatedExecutionHandler, 
#         Portfolio, GfK_long_short
#     )

#     equity = backtest.simulate_trading();
sp500 = pd.read_csv('../../input/long_short_stock/sp500.csv',index_col='Date',parse_dates=['Date'])
#sp500_index = equity_dict['10_90']['equity_curve'].index
sp500_index = sp500.loc[start_date:end_date].index
sp500_compare = sp500['Adj Close'][sp500_index]/sp500['Adj Close'][sp500_index[0]]
```

```python
import seaborn
def heat_map_string_with_pct(df,ax,cmap=None):
    df_cpy = df.copy()
    for col in df_cpy.columns:
        df_cpy[col]=df_cpy[col].str.rstrip('%').astype('float') / 100.0
    df_cpy.index.name = 'Stock Pick'
    df_cpy.columns.name = 'Holding Day'
    seaborn.heatmap(df_cpy,annot=True,linewidths=1,fmt='.2f',cmap = cmap)
    
def heat_map_string_without_pct(df,type_input,fmt_input,ax,cmap=None):
    df_cpy = df.copy()
    for col in df_cpy.columns:
        df_cpy[col]=df_cpy[col].astype(type_input)
    df_cpy.index.name = 'Stock Pick'
    df_cpy.columns.name = 'Holding Day'
    seaborn.heatmap(df_cpy,annot=True,linewidths=1,fmt=fmt_input,cmap = cmap)
#seaborn.heatmap(returns)
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
heat_map_string_without_pct(sharpe,'float','.2f',ax)
plt.title('Sharpe'+name);
plt.savefig('Long_Short_fig/Sharpe'+name)

fig, ax = plt.subplots(figsize=(10,10)) 
cmap = seaborn.cm.rocket_r
heat_map_string_with_pct(max_drawdown,ax,cmap=cmap)
plt.title('Max Drawdown'+name);
plt.savefig('Long_Short_fig/Max_Drawdown'+name)

fig, ax = plt.subplots(figsize=(10,10)) 
returns.index.name = 'Stock Pick'
heat_map_string_with_pct(returns,ax)
plt.title('Returns'+name);
plt.savefig('Long_Short_fig/returns'+name)

fig, ax = plt.subplots(figsize=(10,10)) 
cmap = seaborn.cm.rocket_r
heat_map_string_without_pct(drawdown_duration,'int',"d",ax,cmap=cmap)
plt.title('Drawdown Duration'+name);
plt.savefig('Long_Short_fig/Drawdown_Duration_'+name)
```
